```python
#Lets do splitting here
from sklearn.model_selection import train_test_split
X = df.drop('Default', axis =1)
y = df['Default']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 43)

```

```python
#Let's use embedding layers for remainging object columns
#EmploymentType, MaritalStatus,LoanPurpose
#first we need label encoding these three
id_employment_encoder = LabelEncoder()
X_train['EmploymentType_id'] = id_employment_encoder.fit_transform(X_train[['EmploymentType']])
X_test['EmploymentType_id'] = id_employment_encoder.transform(X_test[['EmploymentType']])

id_marital_encoder = LabelEncoder()
X_train['MaritalStatus_id'] = id_marital_encoder.fit_transform(X_train[['MaritalStatus']])
X_test['MaritalStatus_id'] = id_marital_encoder.transform(X_test[['MaritalStatus']])

id_loan_encoder = LabelEncoder()
X_train['LoanPurpose_id'] = id_loan_encoder.fit_transform(X_train[['LoanPurpose']])
X_test['LoanPurpose_id'] = id_loan_encoder.transform(X_test[['LoanPurpose']])

X_train = X_train.drop(['EmploymentType', 'MaritalStatus', 'LoanPurpose'], axis = 1)
X_test = X_test.drop(['EmploymentType', 'MaritalStatus', 'LoanPurpose'], axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
!pip install tensorflow
from tensorflow.keras.layers import Input, Embedding , Flatten

employment_input = Input(shape = (1, ), name = 'employment_input')
maritial_input = Input(shape = (1, ), name = 'marital_input')
loan_input = Input(shape = (1, ), name = 'loan_input')

employment_input_dim = X_train['EmploymentType_id'].nunique()
maritial_input_dim = X_train['MaritalStatus_id'].nunique()
loan_input_dim = X_train['LoanPurpose_id'].nunique()

employment_embedding = Embedding(input_dim=employment_input_dim, output_dim = 4)(employment_input)
maritial_embedding = Embedding(input_dim=maritial_input_dim, output_dim = 4)(maritial_input)
loan_embedding = Embedding(input_dim = loan_input_dim, output_dim = 4)(loan_input)

employment_vec = Flatten()(employment_embedding)
maritial_vec = Flatten()(maritial_embedding)
loan_vec = Flatten()(loan_embedding)
```

```python
X_train.columns
```

```python
#input vectors of rest columns
input_age = Input(shape = (1, ))
input_income = Input(shape = (1, ))
input_loan_amount = Input(shape = (1, ))
input_credit_score = Input(shape = (1, ))
input_months_employed = Input(shape = (1, ))
input_num_credit_lines = Input(shape = (1, ))
input_interest_rate = Input(shape = (1, ))
input_loan_term = Input(shape = (1, ))
input_dti_ratio = Input(shape = (1, ))
input_education = Input(shape = (1, ))
input_HasMortage = Input(shape = (1, ))
input_HasDependents = Input(shape = (1, ))
input_HasCoSigner = Input(shape = (1, ))
```

```python
# Re-initialize X_train and X_test as DataFrames with all necessary processed features
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Assuming 'X' and 'y' are available from previous steps
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=43)

# Re-apply label encoding for EmploymentType, MaritalStatus, LoanPurpose
id_employment_encoder = LabelEncoder()
X_train['EmploymentType_id'] = id_employment_encoder.fit_transform(X_train['EmploymentType'])
X_test['EmploymentType_id'] = id_employment_encoder.transform(X_test['EmploymentType'])

id_marital_encoder = LabelEncoder()
X_train['MaritalStatus_id'] = id_marital_encoder.fit_transform(X_train['MaritalStatus'])
X_test['MaritalStatus_id'] = id_marital_encoder.transform(X_test['MaritalStatus'])

id_loan_encoder = LabelEncoder()
X_train['LoanPurpose_id'] = id_loan_encoder.fit_transform(X_train['LoanPurpose'])
X_test['LoanPurpose_id'] = id_loan_encoder.transform(X_test['LoanPurpose'])

X_train = X_train.drop(['EmploymentType', 'MaritalStatus', 'LoanPurpose'], axis=1)
X_test = X_test.drop(['EmploymentType', 'MaritalStatus', 'LoanPurpose'], axis=1)

# Now apply StandardScaler to numerical columns
numerical_cols = ['Age','Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed', 'NumCreditLines', 'InterestRate', 'LoanTerm', 'DTIRatio', 'Education', 'HasMortgage', 'HasDependents', 'HasCoSigner']

# Scale numerical columns within the DataFrames
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])
```

```python
with open('/content/drive/MyDrive/Projects/Loan Default Prediction/pickle/scaler.pkl', 'wb') as file:
  pickle.dump(scaler, file)
```

```python
from tensorflow.keras.layers import Concatenate
concatenate_layer = Concatenate()
combined = concatenate_layer([input_age,input_income, input_loan_amount,input_credit_score, input_months_employed, input_num_credit_lines, input_interest_rate, input_loan_term, input_dti_ratio, input_education, input_HasMortage, input_HasDependents, input_HasCoSigner, employment_vec,maritial_vec,loan_vec])
```

```python
from tensorflow.keras.layers import Embedding, Dense, Dropout
x = Dense(128, activation = 'relu')(combined)
x = Dropout(0.2)(x)
x = Dense(64, activation = 'relu')(x)
x = Dropout(0.2)(x)
output = Dense(1, activation = 'sigmoid')(x)
```

```python
from tensorflow.keras.models import Model

model = Model(
    inputs = [input_age,input_income, input_loan_amount,input_credit_score, input_months_employed, input_num_credit_lines, input_interest_rate, input_loan_term, input_dti_ratio, input_education, input_HasMortage, input_HasDependents, input_HasCoSigner, employment_input,maritial_input,loan_input],
    outputs = output
)
```

```python
#we need to compile the model now
from tensorflow.keras.optimizers import Adam
import tensorflow
adam = Adam(learning_rate = 0.001)
model.compile(optimizer = adam, loss = 'binary_crossentropy', metrics=['accuracy', tensorflow.keras.metrics.AUC(name = 'auc')])
```

```python
#Callbacks initialization
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
early_stopping = EarlyStopping(monitor = 'val_loss', patience = 10, restore_best_weights = True)
tensorboard = TensorBoard(log_dir = '/content/sample_data')
```

```python
history = model.fit(
    [X_train['Age'], X_train['Income'], X_train['LoanAmount'], X_train['CreditScore'], X_train['MonthsEmployed'],
     X_train['NumCreditLines'], X_train['InterestRate'], X_train['LoanTerm'], X_train['DTIRatio'], X_train['Education'],
     X_train['HasMortgage'], X_train['HasDependents'], X_train['HasCoSigner'], X_train['EmploymentType_id'],
     X_train['MaritalStatus_id'], X_train['LoanPurpose_id']],
    y_train,
    epochs=50,
    batch_size=32,
    validation_data=(
        [X_test['Age'], X_test['Income'], X_test['LoanAmount'], X_test['CreditScore'], X_test['MonthsEmployed'],
         X_test['NumCreditLines'], X_test['InterestRate'], X_test['LoanTerm'], X_test['DTIRatio'], X_test['Education'],
         X_test['HasMortgage'], X_test['HasDependents'], X_test['HasCoSigner'], X_test['EmploymentType_id'],
         X_test['MaritalStatus_id'], X_test['LoanPurpose_id']],
        y_test
    ),
    callbacks=[early_stopping, tensorboard]
)
```
